```python
from langchain_community.tools import YouTubeSearchTool
from crewai import Task, Agent, Process, Crew
from crewai_tools import SerperDevTool
import yaml
from utils import init_LLM
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
llm = init_LLM(AIML_API_KEY=os.getenv('AIML_API_KEY'))
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(
        base_url=os.getenv('AIML_API_KEY'),
        api_key="https://api.aimlapi.com/v1",    
    )
```

```python
search_tool = SerperDevTool()
files = {
    'agents': '../config/agents.yaml',
    'tasks': '../config/tasks.yaml'
}

# Load configurations from YAML files
configs = {}
for config_type, file_path in files.items():
    with open(file_path, 'r') as file:
        configs[config_type] = yaml.safe_load(file)

# Assign loaded configurations to specific variables
agents_config = configs['agents']
tasks_config = configs['tasks']
```

```python
first_aid_chatbot = Agent(config=agents_config['first_aid_chatbot'], llm=llm)
first_aid_chatbot.tools=[search_tool]

first_aid_response_task = Task(config=tasks_config['first_aid_response_task'], agent=first_aid_chatbot)
first_aid_response_task.tools=[search_tool]
```

```python
crew = Crew(
    agents=[first_aid_chatbot],
    tasks=[first_aid_response_task],
    process=Process.sequential,
)

# Kickoff del Crew
result = crew.kickoff(inputs={"query": "What should I do for a choking adult?"})
```
